### **What is RFM?**
RFM stands for Recency, Frequency, and Monetary value. It's a marketing analysis tool used to identify a company's best customers by examining their transaction history. Here’s a breakdown of each component:

* **Recency (R)**: This metric measures how recently a customer has made a purchase. Customers who have purchased recently are more likely to buy again than customers who have not purchased for a long time. Recency is typically measured by the number of days since the last purchase.

* **Frequency (F)**: This metric measures how often a customer makes a purchase within a specific period. Frequent purchasers are more likely to become repeat customers and are considered more loyal.

* **Monetary value (M)**: This metric measures how much money a customer spends on purchases. Customers who spend more are more valuable to the business.
### **Why Use RFM?**
RFM analysis helps businesses to:

1. **Identify Loyal Customers**: By identifying who buys most frequently and spends the most, businesses can tailor marketing efforts to retain these valuable customers.

2. **Increase Customer Retention**: By focusing on customers who have not made recent purchases but have bought frequently in the past, businesses can target marketing efforts to re-engage these customers.

3. **Optimize Marketing Campaigns**: By segmenting customers based on their RFM scores, businesses can create more targeted marketing campaigns, leading to higher response rates and increased ROI.

4. **Personalize Customer Experience**: Understanding the purchasing behavior of different customer segments allows businesses to tailor their products, services, and communications to meet the needs and preferences of each segment.

### **How to Calculate RFM Scores**
**Step-by-Step Calculation:**
1. **Recency (R):** Determine the time interval (e.g., days, weeks) since the customer's last purchase. The shorter the interval, the more recent the purchase.
2. **Frequency (F):** Count the total number of purchases made by the customer in the specified period.
3. **Monetary (M):** Sum the total amount spent by the customer in the specified period.

```python
today_date = df["Date"].max()

rfm = df.groupby('CustomerID').agg({'Date': lambda Date: (today_date - Date.max()).days,
                                    'InvoiceID': lambda InvoiceID: InvoiceID.nunique(),
                                    'TotalSales': lambda TotalSales: TotalSales.sum()})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

rfm.reset_index(allow_duplicates=False, inplace = True)

rfm
```

### Combining RFM Scores

Each customer is assigned a score for Recency, Frequency, and Monetary value based on their transaction history. These scores are often on a scale from 1 to 5, where 5 indicates the most recent, most frequent, and highest monetary value, respectively. The scores can then be combined to create an overall RFM score, which is used to segment customers into different categories for analysis and targeted marketing.

**Step-by-Step Calculation:** 
* Recency Score: Customers who made purchases more recently receive higher scores (5 is the highest, 1 is the lowest).
* Frequency Score: Customers who made purchases more frequently receive higher scores.
* Monetary Score: Customers who spent more money receive higher scores.

```python
rfm["Recency_score"] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm["Frequency_score"] = pd.qcut(rfm['Frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["Monetary_score"] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])

rfm["RFM_SCORE"] = (rfm['Recency_score'].astype(str) + rfm['Frequency_score'].astype(str))

seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

rfm['Segment'] = rfm['RFM_SCORE'].replace(seg_map, regex=True)

rfm.head()
```

### RFM Segments
Customers are segmented into different groups based on their RFM scores. Here's a common way to interpret the segments:
1. **Hibernating**
* Description: Customers who have made a purchase in the past but have not purchased recently.
* Characteristics: Low Recency, Low Frequency, Low Monetary.
* Strategy: Reactivate these customers with special offers or reminders about new products.
2. **At Risk**
* Description: Customers who used to be active and spend a significant amount but have not purchased recently.
* Characteristics: Low Recency, Moderate to High Frequency, Moderate to High Monetary.
* Strategy: Send re-engagement campaigns with attractive offers to win them back.
3. **Can't Lose Them**
* Description: Previously loyal and high-spending customers who have not purchased in a while.
* Characteristics: Low Recency, High Frequency, High Monetary.
* Strategy: Focus on personalized communications and incentives to bring them back.
4. **About to Sleep**
* Description: Customers who have purchased recently but with low frequency and monetary value.
* Characteristics: Moderate Recency, Low Frequency, Low Monetary.
* Strategy: Encourage them to purchase more frequently with targeted promotions.
5. **Need Attention**
* Description: Customers with an average frequency and monetary value who have not purchased very recently.
* Characteristics: Moderate Recency, Moderate Frequency, Moderate Monetary.
* Strategy: Nurture these customers with personalized communications and incentives to boost their loyalty.
6. **Loyal Customers**
* Description: Customers who purchase often and spend a lot.
* Characteristics: High Frequency, High Monetary.
* Strategy: Reward their loyalty with exclusive offers, early access to new products, or loyalty programs.
7. **Promising**
* Description: Newer customers with potential to become loyal.
* Characteristics: High Recency, Moderate Frequency, Low Monetary.
* Strategy: Encourage repeat purchases with targeted offers and excellent customer service.
8. **New Customers**
* Description: Customers who have made their first purchase recently.
* Characteristics: High Recency, Low Frequency, Low Monetary.
* Strategy: Welcome them with onboarding communications and introductory offers to encourage repeat purchases.
9. **Potential Loyalists**
* Description: Customers who have made multiple purchases recently and show potential to become loyal.
* Characteristics: High Recency, High Frequency, Low Monetary.
* Strategy: Strengthen the relationship with personalized offers and excellent service to convert them into loyal customers.
10. **Champions**
* Description: Best customers who purchase often, spend the most, and have purchased recently.
* Characteristics: High Recency, High Frequency, High Monetary.
* Strategy: Celebrate their loyalty with special rewards, exclusive previews, and maintain a strong relationship.
